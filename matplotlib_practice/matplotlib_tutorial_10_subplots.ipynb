We can see that everytime we do something with our plot we are using the plt object that we imported. That's the pi plot object. We never actually created the pi object ourselves. Like PI = plit() for example 

The way we have been doing it is called "staple" b/c we're importing this pie plot object and it has a current state in terms of figures and axis. The figure is the container holding our plots you can think of that as the whole window holding our code and the axis as the actual plots. So the figure and can have multiple plots. 

We have used plt.gcf() which is get current figure and plt.gca() which is get current axis. Many poeple prefer using the more object oriented approach when working with multiple figures and axis. To do this, we can use the sub-plots method.

WE will recreate the plot we currently have but instead we will do a subplot. 

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('data_sample_for_subplots.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, ax = plt.subplots()

ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')

ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax.legend()

ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

Now we got the same plot as before but now we are using this x axis. Why is this userful? Let's say instead of having all of this data in one plot you wanted it broken out. 

You can just add more axis and do it that way. 

```python
data = pd.read_csv('data_sample_for_subplots.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')


ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

Now you can see we have 1 figure with two different plots. WE can also create 2 different figures. See below

```python
data = pd.read_csv('data_sample_for_subplots.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')


ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
