In this video we will be going over time series data and plotting dates in matplotlib. There is so much data out there that contains date info 
so knowing how to plot this propertly is going to be valuable.

First we will create a basic example and then we will use real world data from a csv file on bitcoin prices

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7] # this is just a list of seven random values. So to plot these dates see code below
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b,%d %Y')

plt.gca().xaxis.set_major_formatter(date_format)


plt.tight_layout()

plt.show()
```

Now let's look at the csv data and plot the bitcoin prices. 

```python
data = pd.read_csv('data_sample_for_timeseries.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date'] # as is this is being read in as text not dates so we need to cast it. Otherwise, this will be read in out of order. The code above casted it
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')


plt.tight_layout()

plt.show()
```
