In this video we are going to look at how to create pie charts. It will show the porportion that each data point will contribute to the whole. 

First we will do a basic exampel to show how this works and then we will use some real world data. 

Resource Link: https://www.youtube.com/watch?v=MPiz50TsyF0&list=PL-osiE80TeTvipOqomVEeZ1HRrcEvtZB_&index=3

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

#let's start off representing 2 numbers in a pie chart
slices = [60, 40]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}) # creates the black line separator.

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Let's change the color of the slices. We create a new list that corresponds to the color that we want for each slice.

```python
plt.style.use("fivethirtyeight")

#let's start off representing 2 numbers in a pie chart
slices = [60, 40,30,20]
labels = ['Sixty', 'Forty','Extra1','Extra2']
colors = ['blue','red','yellow','green'] # you can also input hex color codes to get more precise colors.

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'}) # creates the black line separator.

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Now, let's look at a real world example. 

```python
# Language Popularity
prog_slices = [59219, 55466, 47544, 36443, 35917]
prog_lang = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.style.use("fivethirtyeight")

plt.pie(prog_slices, labels=prog_lang, wedgeprops={'edgecolor': 'black'})

plt.title("Langauges Used")
plt.tight_layout()
plt.show()
```

Let's use the explode arg which will highlight a certain slice for say presentation purposes or something...

```python
plt.style.use("fivethirtyeight")
explode = [0,0,0,0.1,0]

plt.pie(prog_slices, labels=prog_lang, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'})

plt.title("Langauges Used")
plt.tight_layout()
plt.show()
```
