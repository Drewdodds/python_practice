In this vide we will be going over scatter plots. Scatrter plots are great when you wqant to show the relationship between two sets of values and see if they are correlated.

First, we will look at a basic example and then we will look at some real world data. We will look at views and likes on Corey's youtube vids. 

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x_test = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y_test = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]


plt.scatter(x_test,y_test, s=100, c='green', edgecolors='black', linewidth=1, alpha=0.75)
plt.tight_layout()
plt.show()
```

The scatter looks totally random so that likely means there is no relationship between the two variables. We don't see trends here and even that can be important. That there is no trend. 

Let's look at some basic customizations to make it more visually friendly. 

Now, let's look at something different. Let's see how the colors and sizes can actually be on a per mark basis. Rather than applying them to all of the marks so why would you want multiple colors or sizes? Well, having the abilkity to have multiple colors and sizes actually allows us to add additional datasets into our plot. 

Example, we have our current plot above but we wanted to add some additional info. Example, let's pretend that our current plot is some servey data about a bunch of people and we wanted to break down the data further into something more specific. Let's say we had these people rate something from 1 to 10 and we wanted somehow plot their rating as well. We can assign different numbers to these different possibilties and those will then give you different colors on your scatter plot as  long as you pass that into your method. 

See the colors var below. So the numbers in the number list are between 1 and 10. Maybe this coule be a person's answer as to how satisfied they are with a certian product. Each value will correspond to a data point in the x and y variables. Let's pass this in as the color arg.

```python
x_test_2 = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y_test_2 = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
          538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x_test_2,y_test_2, s=sizes, c=colors, cmap='Greens', edgecolors='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
plt.tight_layout()
plt.show()
```

So the stronger colors are a higher rating.
We can also change the sizes of the data points as well. This can add in another way of explaining that data further. I have seen some for populations for example Where the countries with the higher populations have bigger points. 

So you can pass in a size variable to the s arg. See above for the s arg. Now, the higher the rating, the more green it is and the larger.

Now, let's create scatter plot from some real world data. The data is from youtoube API. Let's do a scatter on total views and total likes. And we also have the ratio to likes to dislikes. 


```python
data = pd.read_csv('data_sample_for_scatter.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count,likes, c=ratio,cmap='summer', edgecolors='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')
plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.show()
```

So it is obvious that the more view and vid has the more likes there are. You can see that as the views increase, so does the likes.
